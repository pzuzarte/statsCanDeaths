```python
import pandas as pd
import numpy as np
import plotly.express as px

```

```python
df = pd.read_csv('~/Desktop/canadaStats/Deaths_5x1.tsv', sep='\s+')
```

```python
df.head(30)
```

```python
sum_by_category = df.groupby('Year')['Total'].sum()
sum_by_category
```

```python
df['Sum_by_Category'] = df.groupby('Year')['Total'].transform('sum')
df['Proportion of Deaths'] = df['Total']/df['Sum_by_Category']
df
```

```python
fig = px.bar(df, x="Year", y="Total", color='Age') 
fig.show()
```

```python
fig = px.bar(df, x="Year", y='Proportion of Deaths', color='Age', height=1200, width=1400, title='Proportion of Deaths in Canada by Age from 1922 to 2022') 
fig.update_layout(legend_traceorder="reversed", 
                  title={'text': 'Proportion of Deaths in Canada by Age from 1921 to 2022', 'xanchor':'center', 'yanchor':'top', 'y':0.9, 'x':0.5},
                  xaxis = dict(tickmode = 'linear',dtick = 1),
                  xaxis_title_font_size=20,
                  yaxis_title_font_size=20
                    )
         

fig.update_xaxes(tickangle=90)

fig.show()

fig.write_html("/Users/pzuzarte/Desktop/canadaStats/proportions_plot.html")
```
